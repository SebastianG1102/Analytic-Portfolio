***FORWARD***

```python
from dmba import backward_elimination, forward_selection, stepwise_selection
from dmba import adjusted_r2_score, AIC_score, BIC_score
```

```python
# The initial model is the constant model - this requires special handling
# in train_model and score_model
def resampled_credit_card_forw(variables):
    if len(variables) == 0:
        return None
    credit_card_forw = LinearRegression()
    credit_card_forw.fit(resampled_X, resampled_y)
    return credit_card_forw

def score_credit_card_forw(credit_card_forw, variables):
    if len(variables) == 0:
        return AIC_score(resampled_y, [resampled_y.mean()] * len(resampled_y), credit_card_forw, df=1)
    return AIC_score(resampled_y, credit_card_forw.predict(resampled_X), credit_card_forw)

best_forw_model, best_forw_variables = forward_selection(resampled_X.columns, resampled_credit_card_forw, score_credit_card_forw, verbose=True)

print(best_forw_variables)
```

```python
#Error
```

```python
regressionSummary(valid_y, best_forw_model.predict(valid_X[resampled_X.columns]))
```

***BACKWARDS***

```python
def resampled_credit_card_back(variables):
    credit_card_back = LinearRegression()
    credit_card_back.fit(resampled_X, resampled_y)
    return credit_card_back

def score_credit_card_back(credit_card_back, variables):
    return AIC_score(resampled_y, credit_card_back.predict(resampled_X), credit_card_back)

best_back_credit_card_back, best_back_variables = backward_elimination(resampled_X.columns, resampled_credit_card_back, score_credit_card_back, verbose=True)

print(best_back_variables)
```

```python
regressionSummary(valid_y, best_back_credit_card_back.predict(valid_X[train_X.columns]))
```
